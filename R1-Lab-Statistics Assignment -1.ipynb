```python
def Goal_diff_count():
    sports['Diff'] = sports['GoalsFor'] - sports['GoalsAgainst']
    maxdiff= sports['Diff'].max()
    mindiff= sports['Diff'].min()
    maxteam=sports['Team'][sports['Diff']==maxdiff]
    minteam=sports['Team'][sports['Diff']==mindiff]
    print('Team with max difference',maxteam)
    print('Team with min difference',minteam)
    
    
Goal_diff_count()
print(sports[['Team','Diff']])
```
